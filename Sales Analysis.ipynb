2. ANALYZING THE DATA

```python
#Finding the total sales by branchs
branch_sales = df.groupby('Branch')['Total'].sum().sort_values(ascending=False)
print(branch_sales)
branch_sales.plot(kind='bar', title="Total Sales By branch",figsize=(8,5), colormap='viridis')
plt.ylabel("Total Sales")
plt.show()
```

It shows that Branch C has the highest total sales, while Branch B has the lowest total sales

```python
#Total sales differences of Branch C and B. Highest and lowest
C =  branch_sales.max()
B =  branch_sales.min()
sales_differnce = C - B

C =  branch_sales.idxmax()
B =  branch_sales.idxmin()

print(f"The highest Selling Branch is {C} with sales of {branch_sales.max()}")
print(f"The lowest Selling Branch is {B} with sales of {branch_sales.min()}")
print(f"The total sales difference between them is {sales_differnce}")

```

```python
#Finding gender preference for Each product line
import seaborn as sns
plt.figure(figsize=(8,5))
sns.countplot(x='Product line',hue='Gender', data=df, legend=False, palette='viridis')
plt.xticks(rotation=45)
plt.title("Gender product Preference")
plt.xlabel("Product line")
plt.ylabel("cogs") # Cost of goods
plt.legend(title="Gender")
plt.plot()
```

The BLUE colour is for female while GREEN is for male

```python
gender_preference = df.groupby(['Product line', 'Gender']).size().unstack()
print(gender_preference)
```

```python
dominant_gender = gender_preference.idxmax(axis=1)
print(dominant_gender)
```

It shows that in the Electronic accesories product line it  is mostly the Male gender that purchases the goods sold.

It shows that in the Fashion accesories  product line it  is mostly the Female gender that purchases the goods sold.

It shows that in the Food and beverages  product line it  is mostly the Female gender that purchases the goods sold.

It shows that in the Health and beauty product line it  is mostly the Male gender that purchases the goods sold.

It shows that in the Home and lifesyles product line it  is mostly the Male gender that purchases the goods sold.

It shows that in the Sports and travel  product line it  is mostly the Female gender that purchases the goods sold.
